```python
#Import Pandas, Matplotlib and Numpy to import required data and produce supporting graphs
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
#Get data
df_CTS = pd.read_csv('ConstructionTimeSeriesDataV2.csv')
```

#### Describe below graph:
- For this assigment, I decided to highlight the correlation between private and total construction spending, and how it differs from public spending.
- The below scatter plot highlights the strong positive correlation between the two sets of data. This is indicated by how tighly packed the points are around the best fit line that is also displayed.
- If we want to predict future construction spending, we can use prior trends on private spending as a baseline for our model. 

```python
#Plot data
fig,ax = plt.subplots()
ax.scatter(x = df_CTS['Total Construction'],y = df_CTS['Private Construction'], c = 'g', alpha=0.5,s=75)
ax.xaxis.set_label_text('Total Construction Spending', fontsize = 12)
ax.yaxis.set_label_text('Private Construction Spending', fontsize = 12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

#Format tick labels to currency
formatter = ticker.FormatStrFormatter('$%1.0f')
ax.yaxis.set_major_formatter(formatter)
ax.xaxis.set_major_formatter(formatter)

#Draw best fit line to display strong positive correlation
axes = plt.gca()
m, b = np.polyfit(df_CTS['Total Construction'], df_CTS['Private Construction'], 1)
X_plot = np.linspace(axes.get_xlim()[0], axes.get_xlim()[1], 100)
plt.plot(X_plot, m*X_plot + b, '-', color = 'k', alpha = 0.5)

#Add graph title and adjust size
fig.suptitle('Private vs Total Construction Spending', fontsize = 16)
fig.set_size_inches(16,8)
```

- In contrast, the below graph showing the relationship between public and total construction spending is not as telling. 
- Although a positive trend exists, it is no where near the same level as private spending data. 

```python
#Plot data
fig,ax = plt.subplots()
ax.scatter(x = df_CTS['Total Construction'],y = df_CTS['Public Construction'], c = 'b', alpha=0.5,s=75)
ax.xaxis.set_label_text('Total Construction Spending', fontsize = 12)
ax.yaxis.set_label_text('Public Construction Spending', fontsize = 12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

#Format tick labels to currency
formatter = ticker.FormatStrFormatter('$%1.0f')
ax.yaxis.set_major_formatter(formatter)
ax.xaxis.set_major_formatter(formatter)

#Draw best fit line to display weak positive correlation
axes = plt.gca()
m, b = np.polyfit(df_CTS['Total Construction'], df_CTS['Public Construction'], 1)
X_plot = np.linspace(axes.get_xlim()[0], axes.get_xlim()[1], 100)
plt.plot(X_plot, m*X_plot + b, '-', color = 'k', alpha = 0.5)

#Add graph title and adjust size
fig.suptitle('Public vs Total Construction Spending', fontsize = 16)
fig.set_size_inches(16,8)
```
